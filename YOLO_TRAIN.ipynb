```python
from ultralytics import YOLO
import pandas as pd

model = YOLO("/content/drive/MyDrive/best.pt")


results = model.track(
    source="input.mp4",
    tracker="botsort.yaml",
    save=True,
    save_txt=False,
    stream=True
)


output = []


for frame_idx, r in enumerate(results):
    if r.boxes.id is not None:
        for box, cls, conf, track_id in zip(r.boxes.xyxy, r.boxes.cls, r.boxes.conf, r.boxes.id):
            x1, y1, x2, y2 = box.tolist()
            cx = (x1 + x2) / 2
            cy = (y1 + y2) / 2


            output.append({
                "frame": frame_idx,
                "id": int(track_id),
                "cls": int(cls),
                "conf": float(conf),
                "cx": cx,
                "cy": cy
            })


df = pd.DataFrame(output)
df.to_csv("/content/tracking_data.csv", index=False)
```

```python
df = pd.read_csv("/content/tracking_data.csv")

frame_interval = 1 / 30

speed_data = []

for track_id in df["id"].unique():
    track_df = df[df["id"] == track_id].sort_values("frame")

    if len(track_df) < 2:
        continue

    distances = []
    frames = track_df["frame"].values
    cxs = track_df["cx"].values
    cys = track_df["cy"].values

    for i in range(1, len(track_df)):
        dx = cxs[i] - cxs[i-1]
        dy = cys[i] - cys[i-1]
        dist = (dx**2 + dy**2)**0.5
        distances.append(dist)

    if distances:
        avg_pixel_speed = sum(distances) / (len(distances) * frame_interval)
        speed_data.append({
            "id": track_id,
            "avg_pixel_speed": avg_pixel_speed
        })

speed_df = pd.DataFrame(speed_data)
speed_df.to_csv("/content/avg_pixel_speeds.csv", index=False)
```
